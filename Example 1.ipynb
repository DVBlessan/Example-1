```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sbn
import warnings
warnings.simplefilter('ignore')
```

```python
#1
testdata={'school_code': ['s001','s002','s003','s001','s002','s004'],
 'class': ['V', 'V', 'VI', 'VI', 'V', 'VI'],
 'name': ['Alberto Franco','Gino Mcneill','Ryan Parkes', 'Eesha Hinton', 'GinoMcneill', 'David Parkes'],
 'date_Of_Birth ':['15/05/2002','17/05/2002','16/02/1999','25/09/1998','11/05/2002','15/09/1997'],
 'age': [12, 12, 13, 13, 14, 12],
 'height': [173, 192, 186, 167, 151, 159],
 'weight': [35, 32, 33, 30, 31, 32],
 'address': ['street1', 'street2', 'street3', 'street1', 'street2', 'street4']}
test_data=pd.DataFrame(testdata,index=['S1', 'S2', 'S3', 'S4', 'S5', 'S6'])
```

```python
test_data
```

```python
test_data.info()
```

```python
grouped=test_data.groupby(['school_code'])
```

```python
for name,group in grouped:
    print(name)
    print(group)
```

```python
#2
meanminmax=test_data.groupby('school_code').agg({'age':[np.mean,np.min,np.max]})
print(meanminmax)
```

```python
grouped=test_data.groupby(['school_code','class'])
```

```python
for name,group in grouped:
    print(name)
    print(group)
```

```python
#4
exam_data = {'name': ['Anastasia', 'Dima', 'Katherine', 'James', 'Emily', 'Michael',
'Matthew', 'Laura', 'Kevin', 'Jonas'],
'score': [12.5, 9, 16.5, np.nan, 9, 20, 14.5, np.nan, 8, 19],
'attempts': [1, 3, 2, 3, 2, 3, 1, 1, 2, 1],
'qualify': ['yes', 'no', 'yes', 'no', 'no', 'yes', 'yes', 'no', 'no', 'yes']}
```

```python
sample_data=pd.DataFrame(exam_data,index=['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j'])
```

```python
sample_data
```

```python
#5
sample_data.isna().sum()
```

```python
#6
sample_data.fillna(value=sample_data.mean(),inplace=True)
```

```python
sample_data
```

```python
#7
plt.figure(figsize=(10,9))
plt.bar(sample_data['name'],sample_data['score'])
plt.xlabel('name',size=10)
plt.ylabel('score',size=10)
plt.title("score of the students")
plt.show()
```

```python
plt.figure(figsize=(10,9))
sample_data.qualify.value_counts().plot.pie(labels=['yes','no'],autopct='%1.0f%%')
plt.title("number of students get qualify",size=10)
plt.show()
```

```python
#8
sample_data.to_csv('sample_data.csv')
```
